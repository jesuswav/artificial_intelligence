```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# ------------------------------
# Datos iniciales
# ------------------------------
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 4, 6])
```

```python
# ------------------------------
# 1) Cálculo manual (fórmulas de mínimos cuadrados)
# ------------------------------
n = len(x)

x_mean = np.mean(x)
y_mean = np.mean(y)

# Fórmulas de mínimos cuadrados
b = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)  # pendiente
a = y_mean - b * x_mean  # intercepto

print("Cálculo manual:")
print(f"Pendiente (b): {b:.2f}")
print(f"Intercepto (a): {a:.2f}")


# ------------------------------
# 2) Usando scikit-learn
# ------------------------------
x_reshape = x.reshape(-1, 1)  # sklearn requiere matriz columna
model = LinearRegression()
model.fit(x_reshape, y)

print("\nScikit-learn:")
print(f"Pendiente (b): {model.coef_[0]:.2f}")
print(f"Intercepto (a): {model.intercept_:.2f}")
```

```python

# ------------------------------
# Gráfica
# ------------------------------
plt.scatter(x, y, color="blue", label="Datos")
plt.plot(x, a + b*x, color="red", label="Recta (manual)")
plt.plot(x, model.predict(x_reshape), color="green", linestyle="--", label="Recta (sklearn)")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Regresión Lineal por Mínimos Cuadrados")
plt.legend()
plt.show()
```
